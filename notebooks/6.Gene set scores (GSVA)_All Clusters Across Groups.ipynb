# Render heatmap for the GSVA scores

## No row clustering

```r
library(pheatmap)
options(repr.plot.width=15, repr.plot.height=10)
#pdf(file = "PathwayGSVA_Total.pdf",width=15, height = 13)
xx<-pheatmap(immune.combined.bulk@assays$GSVA@data, cluster_cols = F, cluster_rows=F,show_rownames = T)
#dev.off()

save_pheatmap_pdf <- function(x, filename, width=17, height=13) {
   stopifnot(!missing(x))
   stopifnot(!missing(filename))
   pdf(filename, width=width, height=height)
   grid::grid.newpage()
   grid::grid.draw(x$gtable)
   dev.off()
}

save_pheatmap_pdf(xx, "PathwayGSVA_AllClusters.pdf",width=18, height=12)
```
